```python
import pandas as pd 
import sklearn.metrics as metrics
```

```python
results = open('./exp_result_feed3.csv', 'r')
results = pd.read_csv(results, dtype={'occupancy_truth': str})
results['occupancy_truth']
```

```python
results['occupancy_pred'] = results['occupancy_pred'].apply(lambda x: x.replace(' ', '').replace('[','').replace(']',''))
results
```

```python
pred = ""
truth = []

for index, l in results.iterrows():
    pred+=l['occupancy_pred']
    truth+=l['occupancy_truth']

pred = list(pred)
truth = list(truth)

print(pred)
print(truth)

f1score = 100 * metrics.f1_score(truth, pred, average='weighted')
acc = 100 * metrics.accuracy_score(truth, pred)
# # tn, fp, fn, tp = metrics.confusion_matrix(truth, predicted).ravel()
# false_positive = fp / (fp + tn)
# false_negative = fn / (tp + fn)
# true_positive = tp / (tp + fn)
# true_negative = tn / (tn + fp)

print(f1score)
print(acc)
```
